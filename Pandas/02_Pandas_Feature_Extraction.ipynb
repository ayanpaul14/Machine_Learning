```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
df.loc[2]['Title']
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data == data + i
        if i == '(':
            check = True
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps", "", regex=False)
```

```python
df["Episodes"] = pd.to_numeric(df["Episodes"], errors="coerce")
```

```python
df["Episodes"] = df["Episodes"].fillna(0).astype(int)
```

```python
df.loc[3]['Title']
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+18):
                data += txt[j]

            return data
```

```python
df['Title'] = df['Title'].apply(extraction_time)
```
